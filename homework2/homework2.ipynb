Теперь дозаполним числовые рейтинги футболистов International Reputation, Skill Moves.

```python
df['International Reputation']=df['International Reputation'].fillna(df['International Reputation'].mean())
df['Skill Moves']=df['Skill Moves'].fillna(df['Skill Moves'].mean())
df.info()
```

```python
df['Age Category']=['before 20' if i<20 else 'after 20' for i in df['Age']]
df.sample(10)
```

```python
def calc_age_category(age):
  if age<20:
    return '[0;20)'
  if age>=20 and age<30:
    return '[20;30)'
  if age>=30 and age<=36:
    return '[30;36]'
  if age>36:
    return '>36'
  return 'error: age<0'
```

```python
df['Age Category2']=df['Age'].apply(calc_age_category)
df.sample(10)
```
